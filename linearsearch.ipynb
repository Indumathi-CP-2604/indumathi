```python
import random
import time
import matplotlib.pyplot as plt
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
def generate_list(n):
    return [random.randint(1, 1000000) for _ in range(n)]
def measure_search_time(n, target):
    arr = generate_list(n)
    start_time = time.time()
    linear_search(arr, target)
    end_time = time.time()
    return end_time - start_time
n_values = []
time_taken_values = []
target = random.randint(1, 1000000)
for n in range(100, 10001, 100):
    time_taken = measure_search_time(n, target)
    n_values.append(n)
    time_taken_values.append(time_taken)
plt.plot(n_values, time_taken_values, marker='o')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('Linear Search: Time Taken vs n')
plt.grid(True)
plt.show()
```
