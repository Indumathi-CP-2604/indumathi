```python
import random
import time
import matplotlib.pyplot as plt

# Function to implement linear search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1  # Return -1 if the element is not found

# Function to generate a list of random integers
def generate_list(n):
    return [random.randint(1, 1000000) for _ in range(n)]

# Function to measure time taken for linear search
def measure_search_time(n, target):
    arr = generate_list(n)  # Generate a list of size n
    start_time = time.time()  # Record the start time
    linear_search(arr, target)  # Perform linear search
    end_time = time.time()  # Record the end time
    return end_time - start_time  # Return the time taken

# Experiment: Vary n and record time taken
n_values = []
time_taken_values = []

# Choose a target element to search for
target = random.randint(1, 1000000)

# Repeat the experiment for different values of n (list size)
for n in range(100, 10001, 100):  # List sizes from 100 to 10,000
    time_taken = measure_search_time(n, target)
    n_values.append(n)
    time_taken_values.append(time_taken)

# Plotting the graph
plt.plot(n_values, time_taken_values, marker='o')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('Linear Search: Time Taken vs n')
plt.grid(True)
plt.show()
```
