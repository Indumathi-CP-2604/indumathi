```python
import time
import random
import matplotlib.pyplot as plt
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2
    if left < n and arr[left] > arr[largest]:
        largest = left
    if right < n and arr[right] > arr[largest]:
        largest = right
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
def heap_sort(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
def measure_time(sort_func, arr):
    start_time = time.time()
    sort_func(arr.copy())
    end_time = time.time()
    return end_time - start_time
def generate_random_list(n):
    return [random.randint(0, 1000) for _ in range(n)]
def run_experiment(n_values):
    insertion_times = []
    heap_times = []
    for n in n_values:
        arr = generate_random_list(n)
        insertion_times.append(measure_time(insertion_sort, arr))
        heap_times.append(measure_time(heap_sort, arr))
    return insertion_times, heap_times
if __name__ == "__main__":
    n_values = [10, 50, 100, 200, 500, 1000, 2000, 3000, 4000, 5000]
    insertion_times, heap_times = run_experiment(n_values)
    plt.plot(n_values, insertion_times, label="Insertion Sort")
    plt.plot(n_values, heap_times, label="Heap Sort")
    plt.xlabel("Number of Elements (n)")
    plt.ylabel("Time Taken (seconds)")
    plt.title("Sorting Time vs. Number of Elements")
    plt.legend()
    plt.grid(True)
    plt.show()
```
